```python
import os
import numpy as np
import cv2
import matplotlib.pyplot as plt
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']
%matplotlib inline
```

```python
#图片矢量化
def loadimage(path_to_image):
    image = cv2.imread(path_to_image, 0) #读取图片
    image_vector = image.reshape((image.size, 1)).copy() # 注意：np.copy() 是必要的么？
    return image, image_vector
```

```python
database_path = "ORL"
people_index = 8
sample_index = 5
path_to_image = os.path.join(database_path, 's' + str(people_index) + '_' + str(sample_index) + '.bmp')

#读取图片并进行矢量化
image, image_vector = loadimage(path_to_image)

print(image.size, image.shape)
plt.imshow(image)
```

```python
print(image_vector.size, image_vector.shape)
plt.imshow(image_vector)
```

```python
plt.imshow(image_vector.reshape(image.shape))
```
